```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('../project_data/train.csv') 
```

```python
df.head()
```

```python
df.describe()
```

```python
df.info()
```

```python
print(df.isnull().sum())
```

```python
corr_matrix = df.corr()
plt.figure(figsize=(15, 10))
sns.heatmap(corr_matrix)
plt.title('Correlation Matrix')
plt.show()
```

```python
plt.figure(figsize=(10, 5))
sns.histplot(df['SalePrice'], kde=True)
plt.title('Distribution of SalePrice')
plt.show()
```

```python
plt.figure(figsize=(10, 5))
sns.scatterplot(x=df['GrLivArea'], y=df['SalePrice'])
plt.title('Relationship between GrLivArea and SalePrice')
plt.show()
```

```python
plt.figure(figsize=(10, 5))
sns.boxplot(x=df['OverallQual'], y=df['SalePrice'])
plt.title('Relationship between OverallQual and SalePrice')
plt.show()
```

```python
plt.figure(figsize=(10, 5))
sns.boxplot(x='GarageCars', y='SalePrice', data=df)
plt.title('SalePrice vs GarageCars')
plt.show()
```

```python
plt.figure(figsize=(10, 5))
sns.scatterplot(x='TotalBsmtSF', y='SalePrice', data=df)
plt.title('SalePrice vs TotalBsmtSF')
plt.show()
```

```python
plt.figure(figsize=(10, 5))
sns.scatterplot(x='1stFlrSF', y='SalePrice', data=df)
plt.title('SalePrice vs 1stFlrSF')
plt.show()
```

```python
plt.figure(figsize=(10, 5))
sns.boxplot(x='FullBath', y='SalePrice', data=df)
plt.title('SalePrice vs FullBath')
plt.show()
```

```python
plt.figure(figsize=(10, 5))
df.groupby('YearBuilt')['SalePrice'].mean().plot()
plt.title('SalePrice Trend over the Years')
plt.xlabel('Year Built')
plt.ylabel('Average SalePrice')
plt.show()
```

```python
plt.figure(figsize=(10, 5))
df.groupby('YearRemodAdd')['SalePrice'].mean().plot()
plt.title('SalePrice Trend over the Years after Remodeling')
plt.xlabel('Year Remodeled')
plt.ylabel('Average SalePrice')
plt.show()
```

```python
plt.figure(figsize=(15, 10))
sns.barplot(y='Neighborhood', x='SalePrice', data=df, ci=None)
plt.title('SalePrice vs Neighborhood')
plt.show()
```
